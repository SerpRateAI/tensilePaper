```python
wfs3 = np.vstack([np.loadtxt(f'wfs/{i}_h0.txt') for i in np.arange(0, 586, 1)])
wfs4 = np.vstack([np.loadtxt(f'wfs/{i}_h1.txt') for i in np.arange(0, 586, 1)])
wfs5 = np.vstack([np.loadtxt(f'wfs/{i}_h2.txt') for i in np.arange(0, 586, 1)])
wfs6 = np.vstack([np.loadtxt(f'wfs/{i}_h3.txt') for i in np.arange(0, 586, 1)])

wfs = [wfs3, wfs4, wfs5, wfs6]
aic = [aics3, aics4, aics5, aics6]
```
